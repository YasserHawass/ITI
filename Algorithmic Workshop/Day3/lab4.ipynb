```python
import networkx as nx

adjacency_dictionary = {'0': ['1', '8'],
                        '1': ['0', '7'],
                        '2': ['3', '7', '5'],
                        '3': ['0', '4'],
                        '4': ['3', '8'],
                        '5': ['2', '6'],
                        '6': ['5'],
                        '7': ['2', '1'],
                        '8': ['0', '4']}
```

```python
graph = nx.Graph(adjacency_dictionary)
nx.draw(graph, with_labels = True)
```

```python
# DFS for networkx graph
for element in list(nx.dfs_edges(graph, '0')):
    print("Element: ", element)
```

```python
# BFS for networkx graph
for element in list(nx.bfs_edges(graph, '0')):
    print("Element: ", element)
```

```python
import matplotlib.pyplot as plt
import networkx as nx

nodes = {
    0: [1, 2],
    1: [0, 5, 3, 6],
    2: [0, 5, 6],
    3: [1, 7],
    4: [7, 6],
    5: [1, 2, 7],
    6: [1, 4, 7],
    7: [3, 4, 5, 6]
}
# THAT"S RUBBISH AND I HATE IT
G = nx.Graph(nodes)

G.add_edge(0, 1, weight=8)
G.add_edge(0, 2, weight=9)
G.add_edge(1, 5, weight=7)
G.add_edge(1, 3, weight=1)
G.add_edge(1, 6, weight=9)
G.add_edge(1, 0, weight=8)
G.add_edge(2, 5, weight=6)
G.add_edge(2, 6, weight=4)
G.add_edge(2, 0, weight=9)
G.add_edge(3, 7, weight=4)
G.add_edge(3, 1, weight=1)
G.add_edge(4, 7, weight=7)
G.add_edge(4, 6, weight=7)
G.add_edge(5, 1, weight=7)
G.add_edge(5, 2, weight=6)
G.add_edge(5, 7, weight=2)
G.add_edge(6, 1, weight=9)
G.add_edge(6, 2, weight=4)
G.add_edge(6, 4, weight=7)
G.add_edge(6, 7, weight=6)
G.add_edge(7, 3, weight=4)
G.add_edge(7, 4, weight=7)
G.add_edge(7, 5, weight=2)
G.add_edge(7, 6, weight=6)

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='skyblue', edge_color='red', node_size=500, width=2, alpha=0.5)
weights = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=weights, font_size=10)
plt.show()
```

```python
for element in nx.all_pairs_dijkstra_path_length(G):
    print(element)
```

```python
distance, path = nx.single_source_dijkstra(G, source=0, target=5)
print(distance)
print(path)
```
