```python
import requests

url = "https://community-open-weather-map.p.rapidapi.com/climate/month"

querystring = {"q":"Cairo"}

headers = {
    'x-rapidapi-host': "community-open-weather-map.p.rapidapi.com",
    'x-rapidapi-key': "8fa45b2a9fmshc8f62f712753bdfp12822bjsncacb3fefe2aa"
    }

response = requests.request("GET", url, headers=headers, params=querystring)

#print(response.text)
```

```python
import pandas as pd
import json

# Load the JSON to a Python list of dictionaries then convert that list of dictionaries into a pandas dataframe
df_from_API = pd.DataFrame(json.loads(response.text)['list'])

df_from_API.head()
```
